# Sentiment Analysis of iPhone Reviews Using BiLSTM

**Objective:** Train a Bidirectional LSTM (BiLSTM) neural network to classify Amazon iPhone reviews as **Positive**, **Neutral**, or **Negative** based on the review title and description.

**Why BiLSTM for text?**  
Text is sequential data - the order of words carries meaning. For example:
- *"I like this phone"* → Positive  
- *"I do NOT like this phone"* → Negative  

A BiLSTM reads the text in **both directions** (left→right and right→left) simultaneously, giving each word full context from both sides. This makes it far better than a SimpleRNN which reads only left→right and forgets earlier words in long sentences.

**Why set random seeds?**  
`np.random.seed(42)` and `tf.random.set_seed(42)` ensure that weight initialisation and data shuffling produce the exact same result every time you run the notebook. Without seeds, results change slightly on each run, making it impossible to reproduce your findings consistently.
